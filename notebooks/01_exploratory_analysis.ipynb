```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
print(df.shape)
print('---')
print(df.dtypes)
print('---')
print(df.isnull().sum())
```

```python
print(df['TotalCharges'].unique()[:20])
print('---')
df[df['TotalCharges'] == ' ']
```

```python
pd.to_numeric(df['TotalCharges'], errors='coerce').isnull().sum()
```

```python
mask = pd.to_numeric(df['TotalCharges'], errors='coerce').isnull()
df[mask]
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
print(df['TotalCharges'].dtype)
```

```python
# Cuantos clientes se fueron vs se quedaron
df['Churn'].value_counts()
```

```python
sns.countplot(data=df, x='Churn')
plt.title('Distribucion de Churn')
plt.show()
```

```python
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Churn por tipo de contrato')
plt.show()
```

```python
sns.histplot(data=df, x='tenure', hue='Churn', bins=30)
plt.title('Churn por antiguedad del cliente')
plt.show()
```

```python
sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Churn por cargo mensual')
plt.show()
```
